```python
import pandas as pd
```

```python
pd.read_csv('./data/CARGA_GLOBAL_DEL_TRABAJO.csv')
```

```python
tabla1 = pd.read_csv('./data/CARGA_GLOBAL_DEL_TRABAJO.csv')
```

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource

output_notebook()
```

```python
group1 = tabla1.groupby('AREA') 
source = ColumnDataSource(group1)
gph = figure(
    plot_height=250, plot_width=750,
    x_range=group1,  
    title='CARGA GLOBAL DEL TRABAJO'
)
gph.vbar(x='AREA', top='HORAS_PROMEDIO_A_LA_SEMANA_MUJERES_mean', width=0.5, line_color="white", source=source)
gph.xgrid.grid_line_color = None
gph.xaxis.axis_label = "AREA - HORAS PROMEDIO SEMANAL MUJERES"
gph.y_range.start = 0

show(gph)
```

```python
group1 = tabla1.groupby('AREA') 
source = ColumnDataSource(group1)
gph = figure(
    plot_height=250, plot_width=750,
    x_range=group1,  
    title='CARGA GLOBAL DEL TRABAJO' 
)
gph.vbar(x='AREA', top='HORAS_PROMEDIO_A_LA_SEMANA_\nHOMBRES_mean', width=0.5, line_color="black", color = "pink", source=source)
gph.xgrid.grid_line_color = None
gph.xaxis.axis_label = "AREA - HORAS PROMEDIO SEMANAL HOMBRES"
gph.y_range.start = 0

show(gph)
```
